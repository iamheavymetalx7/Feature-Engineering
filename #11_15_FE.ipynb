# **7.Probability ration encoding**
1. Probability of survived based on cabin -- Categorical feature
2. Probability of died -- 1- survived
3. pr(survived)/pr(died)
4. dictionary to map cabin with probability
5. replace with the categorical feature

```python
import pandas as pd
#titanic dataset
```

```python
df=pd.read_csv('train.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
##replacing NaN 
df['Cabin'].fillna('Missing',inplace=True)
df.head(20)
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
prob_df = df.groupby(['Cabin'])['Survived'].mean()
prob_df
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
 prob_df['Died']=1-prob_df['Survived']
 prob_df.head()
```

```python
  prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
  prob_df.head()
```

```python
prob_encoded = prob_df['Probability_ratio'].to_dict()
prob_encoded
```

```python
df['Cabin_encoded']=df['Cabin'].map(prob_encoded)
df.head(20)
```
